# Open Interest — Raw Data Explorer

```python
import pandas as pd

btc = pd.read_parquet('../data/btc_open_interest.parquet')
eth = pd.read_parquet('../data/eth_open_interest.parquet')
```

## Coverage

Data is 5-min resolution, last 90 days.

```python
for name, df in [('BTC', btc), ('ETH', eth)]:
    print(f"{name}: {len(df)} rows  |  {df['timestamp'].min()} → {df['timestamp'].max()}")
```

## Distribution

```python
btc[['open_interest', 'open_interest_value']].describe()
```

## Daily averages

```python
btc_daily = btc.set_index('timestamp').resample('D')[['open_interest', 'open_interest_value']].mean()
btc_daily
```

## BTC vs ETH OI value side by side

```python
btc_d = btc.set_index('timestamp').resample('D')['open_interest_value'].mean().rename('btc_oi_usd')
eth_d = eth.set_index('timestamp').resample('D')['open_interest_value'].mean().rename('eth_oi_usd')
pd.concat([btc_d, eth_d], axis=1)
```
