## During crypto backtest period (2020+)

```python
vix_crypto = vix[vix['timestamp'] >= '2020-01-01'].copy()
vix_crypto['vix'].describe()
```

## Regime thresholds

VIX > 25 = high uncertainty (used in regime split in `6_regime.py`)

```python
vix_crypto['regime'] = vix_crypto['vix'].apply(lambda x: 'high (>25)' if x > 25 else 'low (≤25)')
vix_crypto['period'] = vix_crypto['timestamp'].apply(lambda x: 'pre-2021' if x < pd.Timestamp('2021-01-01') else 'post-2021')
vix_crypto.groupby(['period', 'regime']).size().unstack()
```

## Extreme VIX days

```python
print("Highest VIX days (all time):")
display(vix.nlargest(10, 'vix'))

print("\nHighest VIX days (2020+):")
display(vix_crypto.nlargest(10, 'vix')[['timestamp', 'vix']])
```
